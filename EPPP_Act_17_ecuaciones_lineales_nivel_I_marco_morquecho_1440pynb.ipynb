#### problema 1

Enunciado: $x + 8 = 15$Planteamiento: Tenemos un número al que se le suma 8 y da 15.Despeje: El 8 positivo pasa al otro lado restando.$$x = 15 - 8$$Solución: $x = 7$

```python
# Ejercicio 1
# total = 15
# sumando = 8
total=int(input("Ingrese el total: "))
sumando=int(input("Ingrese el sumando: "))
x = total - sumando
print(f"Solución Ejercicio 1: x = {x}")
```

# Ejercicio 3
Enunciado: $3x + 4x = 21$Planteamiento: Tres veces un número más cuatro veces el mismo número es 21.Despeje: Sumamos los términos con variable ($3 + 4 = 7$).$$7x = 21$$Luego, el 7 que multiplica pasa dividiendo:$$x = 21 / 7$$Solución: $x = 3$

```python
# Ejercicio 3
# coeficiente_total = 3 + 4
# resultado = 21
coeficiente_total = int(input("Ingrese el coeficiente total: "))
resultado = int(input("Ingrese el resultado: "))
x = resultado / coeficiente_total
print(f"Solución Ejercicio 3: x = {int(x)}")
```

#ejercicio 5

Enunciado: $6a - 2a + a = 20$Planteamiento: Operamos con la variable $a$. Recuerda que $a$ sola equivale a $1a$.Despeje: Agrupamos: $(6 - 2 + 1)a = 20$.$$5a = 20$$El 5 pasa dividiendo:$$a = 20 / 5$$Solución: $a = 4$

# Ejercicio 7
Enunciado: $4x + 3 = 2x + 11$Planteamiento: Hay variables y números en ambos lados.Despeje: 1. Llevamos las $x$ a la izquierda: $4x - 2x = 2x$.2. Llevamos los números a la derecha: $11 - 3 = 8$.$$2x = 8$$3. Dividimos: $x = 8 / 2$.Solución: $x = 4$

```python
# Ejercicio 7
# (4x - 2x) = (11 - 3)
# coef_x = 4 - 2
# nums = 11 - 3
coef_x = int(input("Ingrese el coeficiente de x: "))
nums = int(input("Ingrese el número: "))
x = nums / coef_x
print(f"Solución Ejercicio 7: x = {int(x)}")
```

# Ejercicio 9
Enunciado: $9m + 4 = 5m + 20$Planteamiento: Similar al anterior, agrupamos términos semejantes.Despeje:Restamos las $m$: $9m - 5m = 4m$.Restamos los números: $20 - 4 = 16$.$$4m = 16$$Dividimos: $m = 16 / 4$.Solución: $m = 4$

```python
# Ejercicio 9
# (9m - 5m) = (20 - 4)
# coef_m = 9 - 5
# nums = 20 - 4
coef_m = int(input("Ingrese el coeficiente de m: "))
nums = int(input("Ingrese el número: "))
m = nums / coef_m
print(f"Solución Ejercicio 9: m = {int(m)}")
```
