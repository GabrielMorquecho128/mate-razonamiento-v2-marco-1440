#### problema 1
La suma de un número más 9 es igual a 21Paso 1: Análisis y resoluciónPara resolver esto, debemos asignar una variable a nuestra incógnita (el número que no conocemos).Definir la variable: Sea $x$ el número desconocido.Plantear la ecuación: El enunciado dice "un número ($x$) más ($+$) 9 es igual ($=$) a 21".$$x + 9 = 21$$Despejar la variable: Para dejar a la $x$ sola, el 9 que está sumando pasa al otro lado de la igualdad restando.$$x = 21 - 9$$Resolver:$$x = 12$$Resultado: El número es 12.


```python
# Problema: x + 9 = 21

# Valores conocidos
# resultado_total = 21
# numero_sumado = 9
resultado_total=int(input("ingrese el resultado total: "))
numero_sumado=int(input("ingrese el numero sumado: "))
# Despeje: x = resultado_total - numero_sumado
x = resultado_total - numero_sumado

print(f"El número buscado es: {x}")
```

#### problema 2
Las edades de Pedro y AnaPaso 1: Análisis y resoluciónAquí tenemos dos incógnitas, pero podemos expresar una en función de la otra.Definir variables:Edad de Ana: $a$Edad de Pedro: $2a$ (porque es el doble que la de Ana).Plantear la ecuación: La suma de ambas edades es 36.$$a + 2a = 36$$Simplificar: Sumamos los términos semejantes ($1a + 2a = 3a$).$$3a = 36$$Despejar la variable: El 3 que multiplica a la $a$ pasa dividiendo al 36.$$a = \frac{36}{3}$$Resolver:$a = 12$ (Edad de Ana)$2a = 2(12) = 24$ (Edad de Pedro)Resultado: Ana tiene 12 años y Pedro tiene 24 años.

```python
# Problema: a + 2a = 36  ->  3a = 36

# Suma total de las edades
# total_edades = 36
total_edades=int(input("ingrese la edad total: "))
# Proporción total (1 parte de Ana + 2 partes de Pedro = 3 partes)
proporcion_total = 3

# Calculamos la edad de Ana (a)
edad_ana = total_edades / proporcion_total

# Calculamos la edad de Pedro (2a)
edad_pedro = edad_ana * 2

print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
