#### problema 3
: Un número al que se le resta 7 da como resultado 18Paso 1: Análisis y resoluciónDefinir la variable: Sea $x$ el número desconocido.Plantear la ecuación: "Un número ($x$) al que se le resta ($-$) 7 es igual ($=$) a 18".$$x - 7 = 18$$Despejar: El 7 que está restando pasa al otro lado sumando.$$x = 18 + 7$$Resolver: $x = 25$.

```python
# Problema: x - 7 = 18
# resultado = 18
# resta = 7
resultado=int(input("ingrese el resultado: "))
resta=int(input("ingrese la resta: "))
# Despeje: x = resultado + resta
x = resultado + resta
print(f"El número buscado en el problema es: {x}")
```

#### problema 4

El triple de un número más 4 es igual a 31Paso 1: Análisis y resoluciónDefinir la variable: Sea $x$ el número.Plantear la ecuación: "El triple de un número ($3x$) más ($+$) 4 es igual ($=$) a 31".$$3x + 4 = 31$$Despejar:Primero movemos el 4: $3x = 31 - 4 \rightarrow 3x = 27$Luego movemos el 3 (que multiplica) dividiendo: $x = \frac{27}{3}$Resolver: $x = 9$.

```python
# Problema: 3x + 4 = 31
# total = 31
# suma = 4
# multiplicador = 3
tota=int(input("ingrese el total: "))
suma=int(input("ingrese la suma: "))
multiplicador=int(input("ingrese el multiplicador: "))
# Paso a paso:
# 1. Restar el término independiente: 3x = total - suma
# 2. Dividir por el coeficiente: x = (total - suma) / multiplicador
x = (total - suma) / multiplicador
print(f"El número buscado en el problema 4 es: {int(x)}")
```

#### problema 5
Terreno (largo mide 5m más que el ancho, perímetro = 50)Paso 1: Análisis y resoluciónDefinir variables:Ancho: $w$Largo: $w + 5$ (mide 5 metros más que el ancho).Plantear la ecuación: El perímetro de un rectángulo es la suma de todos sus lados: $2(ancho) + 2(largo) = Perímetro$.$$2(w) + 2(w + 5) = 50$$Simplificar:$2w + 2w + 10 = 50$$4w + 10 = 50$Despejar:$4w = 50 - 10 \rightarrow 4w = 40$$w = \frac{40}{4} \rightarrow w = 10$Resolver:Ancho: 10 metros.Largo: $10 + 5 = 15$ metros.

```python
# Problema: 4w + 10 = 50
# perimetro = 50
# diferencia_lados = 5 # El largo es 5m más
perimetro =int(input("ingrese el perimetro: "))
diferencia_lados=int(input("ingrese la diferencia de lados: "))
# Despeje derivado de: 2w + 2(w + 5) = 50 -> 4w + 10 = 50
ancho = (perimetro - (2 * diferencia_lados)) / 4
largo = ancho + diferencia_lados

print(f"El ancho del terreno es: {int(ancho)}m")
print(f"El largo del terreno es: {int(largo)}m")
```
